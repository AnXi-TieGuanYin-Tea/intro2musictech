**✎ 使用模块**

无论是Python本身内置模块，还是之前我们通过`pip`安装好的第三方模块（例如`librosa`）

只要安装完毕，这些模块就可以立刻通过`import`来调用

由于我们需要通过`matplotlib.pyplot`显示图像，用`IPython.display`播放音频

还将用到`librosa`里的方法来加载一段音频，以及`librosa.display`画波形

以上都需要被调用：

```python
%matplotlib inline              
import matplotlib.pyplot as plt  # 将该模块重命名为plt被调用
import IPython.display as ipd    # 将该模块重命名为ipd被调用
import librosa, librosa.display  # 可用逗号分隔被调用的模块
```

**✎ 加载音频文件**

通过`librosa`模块里的`load`方法，我们可以加载指定路径下的音频，如`attachment`文件夹内的`cat-meow.mp3`

该方法会返回两个变量：

- `x`代表音频数据本身
- `sr`代表以何种采样率得到的数据`x`

```python
x, sr = librosa.load('attachment/cat-meow.mp3')
```

如果出现无法加载的问题，极大可能是因为你的电脑需要另外安装[ffmpeg](https://www.ffmpeg.org/)

Linux和OSX系统下的conda会默认安装ffmpeg，但是Windows用户需要额外安装 

**✎ 播放音频**

通过`ipd`模块里的`Audio`方法可以播放采样率为`sr`的音频数据`x`，听到一声猫咪叫~

```python
ipd.Audio(x, rate=sr)
```

**✎ 画出音频波形**

最后我们再用`plt`的`figure`方法规定画出一个尺寸为长15高5的图像

图像是猫咪音频的波形图，即一维数据`x`在不同时间点上的大小,并用`alpha`变量的值规定图像透明度：

```python
plt.figure(figsize=(15, 5))
librosa.display.waveplot(x, sr, alpha=0.8)
```
